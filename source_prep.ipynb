## This is source preparation notebook
###### We shall prepare a small data as a source system, so we can use in spark declarative pipeline
###### We shall have one transaction table (sales) and two lookup tables (stores and products)

```python
# Create a transaction table sales
%sql
Create table if not exists sdp.source.sales
(
  order_id int,
  product_id int,
  revenue float,
  date Date,
  store_id int,
  order_status int,
  comments varchar(1000)
);

INSERT INTO sdp.source.sales VALUES
(1, 1001, 100.00, '2022-01-01', 1, 1, 'this order is good'),
(2, 1002, 200.00, '2022-01-02', 1, 1, 'this order is bad'),
(3, 1003, 300.00, '2022-01-03', 2, 1, 'this order is good'),
(4, 1004, 400.00, '2022-01-04', 2, 0, 'this order is bad');

SELECT * from sdp.source.sales;
```

```python
# Create a lookup table products
%sql
create table if not exists sdp.source.products
(
  product_id int,
  product_name string,
  category string
);

INSERT INTO sdp.source.products VALUES
(1001, 'Product1', 'Category1'),
(1002, 'Product2', 'Category2'),
(1003, 'Product3', 'Category1'),
(1004,  'Product4', 'Category3');
    
SELECT * from sdp.source.products;
```

```python
# Create a lookup table stores
%sql
create table if not exists sdp.source.stores
(
  store_id int,
  store_name string
);

INSERT INTO sdp.source.stores VALUES
(1, 'Store1'),
(2, 'Store2');
    
SELECT * from sdp.source.stores;
```
